```python
class Place:
    def __init__(company, name, city , year_of_start):
        company.name = name
        company.city = city
        company.year_of_start = year_of_start
```

```python
Mercedes = Place("Mercedes", "Berlin", 2024)
print(Mercedes)
print("%s is company in %s which started in the year %d"% (Mercedes.name,Mercedes.city,Mercedes.year_of_start))
```

```python
class Info:
    def __init__(self , name , field , place , year_of_start):
        self.name = name
        self.field = field
        self.place = place
        self.year_of_start = year_of_start
    
    def years(self , year_of_complete):
        return year_of_complete - self.year_of_start
    
    def __str__(self):
        return "%s studied %s in %s in the year %d ."%(self.name , self.field , self.place , self.year_of_start)
    
ruby = Info("Shreyas","MS","Germany",2023)
print(ruby)
print(ruby.years(2023))
        
```

```python
class Person:
    def __init__(self, name, surname, year_of_birth):
        self.name = name
        self.surname = surname
        self.year_of_birth = year_of_birth
    
    def age(self, current_year):
        return current_year - self.year_of_birth
    
    def __str__(self):
        return "%s %s was born in %d ." % (self.name, self.surname, self.year_of_birth)
    
ruby = Person("Shreyas", "T C", 1999)
print(ruby)
print(ruby.age(1999))

class Student(Person):
    def __init__(self, student_id, *args, **kwargs):
        super(Student, self).__init__(*args, **kwargs)
        self._student_id = student_id
        
s = Student(3, 'Shreyas', 'TC',1999 )
print(s)
print(type(s))
print(isinstance(s, Person))
print(isinstance(s, object))
```

```python
class Student(Person):
    def __init__(self, student_id, *args, **kwargs):
        super(Student, self).__init__(*args, **kwargs)
        self._student_id = student_id
        
    def __str__(self):
        return super(Student, self).__str__() + " And has Student ID: %d" % self._student_id
        
s = Student(3, 'shreyas', 'TC', 1999)
print(s)
```
